```python
import sklearn
import pandas as pd
import numpy as np
import datetime
```

```python
print(f"sklearn v: {sklearn.__version__}")
print(f"numpy v:   {np.__version__}")
print(f"pandas v:   {pd.__version__}")
```

```python
automobile_df = pd.read_csv("cars.csv")
automobile_df.head()
```

```python
automobile_df.shape
```

```python
automobile_df = automobile_df.replace('?', np.nan)  # zamieniamy znak ? na NaN żeby pandas to dalej ogarnęła
automobile_df.head()
```

```python
automobile_df.isna().sum()
```

```python
automobile_df['MPG'] = automobile_df['MPG'].fillna(automobile_df['MPG'].mean())  # uzupełniamy NaN w kolumnie MPG (miles per galon) na wyliczoną średnią z pozostałych wartości
automobile_df.isna().sum()
```

```python
automobile_df.head()
```

```python
automobile_df.isnull().sum()  # dalej sprawdzamy kompletność danych
```

```python
automobile_df = automobile_df.dropna()  # usuwamy wiersze z niepełnymi danymi, tutaj akurat można sobie na to pozwolić bo próbka jest dość duża
```

```python
automobile_df.isna().sum()
```

```python
automobile_df.drop(['bore', 'stroke', 	'compression-ratio'], inplace=True, axis=1)  # usuwamy kolumny które nie są nam potrzebne w analizie
automobile_df.head()
```

```python
automobile_df.Year.str.isnumeric().value_counts()  # sprawdzamy czy dane są tego samego typu, czy są muneryczne
```

```python
automobile_df.loc[automobile_df['Year'].str.isnumeric() == False]  # pokazujemy te dane które nie są numeryczne żeby zobaczyć co możemy z tym dalej zrobić
```

```python
extr = automobile_df['Year'].str.extract(r'^(\d{4})', expand=False)  # wycinamy pierwsze 4 cyfry żeby mieć rok produkcji, później podmienimy kolumny
extr.head()
```

```python
automobile_df['Year'].isnull().values.any()  # sprawdzamy czy w tej kolumnie są jeszcze jakieś wartości nienumeryczne
```

```python
automobile_df['Year'] = pd.to_numeric(extr)  # podmieniamy kolumnę Year na extr
automobile_df.head()
```

```python
automobile_df['Cylinders'].str.isnumeric().value_counts()  # zaczynamy czyszczenie kolumny Cylinders, sprawdzamy ile wpisaów nie jest numerycznych
```

```python
automobile_df.loc[automobile_df['Cylinders'].str.isnumeric() == False]  # to pokaże całą ramkę
automobile_df['Cylinders'].loc[automobile_df['Cylinders'].str.isnumeric() == False]  # to pokaże tylko kolumnę Cylinders
```

```python
cylinders = automobile_df['Cylinders'].loc[automobile_df['Cylinders'] != '-']
cmean = cylinders.astype(int).mean()
```

```python
automobile_df.Cylinders = automobile_df.Cylinders.replace('-', cmean).astype(int)
automobile_df.dtypes
```

```python
automobile_df.loc[automobile_df['Displacement'].str.isnumeric() == False]
```

```python
automobile_df.Displacement = pd.to_numeric(automobile_df.Displacement, errors='coerce')  # errors='coerce' it will replace all non-numeric values with NaN.
automobile_df.dtypes
```

```python
automobile_df.drop(['Model'], inplace=True, axis=1)  # axis=1 to kolumny, gdyby było 0 to wiersze, inplace=True - usuwa z oryginalnego datasetu, inplace=False (domyślny) usuwa z ramki danych
automobile_df.Weight = pd.to_numeric(automobile_df.Weight, errors='coerce')
automobile_df.Acceleration = pd.to_numeric(automobile_df.Acceleration, errors='coerce')
automobile_df.head()
```

```python
automobile_df.dtypes
```

```python
automobile_df.Origin.unique()
```

```python
automobile_df.Origin = np.where(automobile_df.Origin.str.contains('US'), 'US', automobile_df.Origin)
automobile_df.head()
```

```python
automobile_df.Origin.unique()
```

```python
automobile_df.Origin = np.where(automobile_df.Origin.str.contains('Japan'), 'Japan', automobile_df.Origin)
automobile_df.Origin = np.where(automobile_df.Origin.str.contains('Europe'), 'Europe', automobile_df.Origin)
automobile_df.Origin.unique()
```

```python
automobile_df.describe()
```

```python
automobile_df['Age'] = datetime.datetime.now().year - automobile_df.Year
automobile_df.drop(['Year'], axis=1, inplace=True)
automobile_df.sample(5)
```

```python
automobile_df.to_csv("cars_processed.csv", index=False)
```
